There are three contestants in a diving competition. In this competition, they will finish three rounds of dives, each with a difficulty value. 1. remove the maximum and minimal of the five scores.
2. calculation the average score of the remaining three scores
3. the effective score is the difficulty multiply the average score in #2.

After all the constestants finish all their rounds of dives, we sum all effective scores of a constestant and calculate his final score. The constestant's rank will be determined by his final score - the constestant with the highest score will have rank 1, and etc..
